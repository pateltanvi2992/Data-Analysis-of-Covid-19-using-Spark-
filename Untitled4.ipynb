```python
import pandas as pd
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df1 = pd.read_csv(url)
```

```python
df=df1[df1['location']=="United States"]
```

```python
df
```

```python
USA_total_cases=df['new_cases'].sum()
```

```python
USA_total_cases
```

```python
USA_total_cases/331420000.0 * 100
```

```python
USA_Fully_people_vaccinated= df['people_fully_vaccinated'].max()
```

```python
USA_Fully_people_vaccinated
```

```python
USA_people_vaccinated= df['people_vaccinated'].max()
```

```python
USA_population= 331420000.0
USA_people_vaccinated_percentage=USA_people_vaccinated/USA_population * 100 
USA_people_vaccinated_percentage
```

```python
USA_total_left_population= 100-USA_people_vaccinated_percentage
USA_total_left_population
```

```python
import matplotlib.pyplot as plt
labels = 'United State population without vaccines', 'people vaccinated till 26th March 2021',
sizes = [USA_total_left_population, USA_people_vaccinated_percentage]
explode = (0, 0.1)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
